# Feature Extraction

```python
from urllib.parse import urlparse, urlencode
import ipaddress
import re
```

## Getting Domain

```python
def getDomain(url):  
    domain = urlparse(url).netloc
    if re.match(r"^www.",domain):
        domain = domain.replace("www.","")
    return domain
```

## IP address

```python
def IP_add_usage(url):
    try:
        ipaddress.ip_address(url)
        return 1 # Phishing
    except:
        return 0  # Normal
```

## Unfamiliar Domain

```python
def unfamilar_domain(url):
    tld_pattern = r'\.([a-zA-Z]{2,})$'

    match = re.search(tld_pattern, url)

    if match:
        tld = match.group(1).lower()  # Access the first capturing group and convert to lowercase
        unknown_tlds = ['xyz', 'buzz', 'cn', 'fn', 'buzz', 'to', 'pw', 'club', 'top', 'tk', 'ws', 'ga', 'gq', 'abc']
        if tld not in unknown_tlds:
            return 0  # Normal
    return  1  # If no TLD is found, consider it abnormal
```

## Unappropriate Signs

```python
def unapp_signs(url):
    signs = ['@', '!', '$', '^', '~', '{', '}']
    for sign in signs:
        if sign in url:
            return 1  # Phishing
    return 0  # Normal
```

## Length of URL

```python
def len_web_add(url):
    if len(url)<75:
        return 1 # Phishing
    else:
        return 0 # Normal
```

## Redirection using '//'

```python
def redirection(url):
    re = url.rfind('//')
    if re > 6:
        if re > 7:
            return 1 # Phishing
        else:
            return 0 # Normal
    else:
        return 0   # Normal
```

## Typo Squatting Error

## "http/https" in Domain name

```python
def http_extra(url):
    domain = urlparse(url).netloc
    if 'https' in domain:
        return 1 #Phishing
    else:
        return 0  #Normal
```

## Shortend URL

```python
def is_shortened_url(url):
    shortening_services = ['bit.ly','tinyurl.com', 't.co', 'goo.gl', 'ow.ly', 'bit.do', 'b.link', 'short.io', 'snip.ly', 'buff.ly']
    domain = urlparse(url).netloc
    
    if len(domain)<9:
        if domain in shortening_services:
            return 0  # Phishing 
        else:
            return 1  # Normal 
    else: 
        return 0 # Normal
    
```

## Checking Web Trafficking

```python
from bs4 import BeautifulSoup
import urllib.request 
import requests

def web_traffic_report(url):
    website = requests.get('https://www.domcop.com/top-10-million-websites')
    try:
        soup = BeautifulSoup(urllib.request.urlopen(website +url).read(),'xml')
        web_rank = soup.find("Rank").text
        web_rank_int = int(web_rank.replce(',',''))
    except TypeError:
        return 1

    if web_rank_int< 10000:
        return 0   # Phishing
    else:
        return 1    #Normal
```

## Pop up detection

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def detect_popups(url):
    options = webdriver.ChromeOptions()
    options.add_argument("--disable-popup-blocking")
    driver = webdriver.Chrome()
    # Wait for the popups to appear
    WebDriverWait(driver, 5).until(EC.number_of_windows_to_be(3))  # Assuming there are 3+ popups
    popup_handles = driver.window_handles
    num_popups = int(len(popup_handles) - 1)  # Subtract 1 for the main window
    
    driver.quit()
    
    if num_popups > 2:
        return 1
    else:
        return 0
```

# Computing URL Features

```python
def features_ex(url,label):
    features = []
    features.append(getDomain(url))
    features.append(IP_add_usage(url))
    features.append(unfamilar_domain(url))
    features.append(unapp_signs(url))
    features.append(len_web_add(url))
    features.append(redirection(url))
    #features.append(is_typosquatting(url))
    features.append(http_extra(url))
    features.append(is_shortened_url(url))
    #features.append(web_traffic_report(url))
    #features.append(detect_popups(url))
    
    
    return features
```
